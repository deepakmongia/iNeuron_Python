### 1. You are given two strings s and t. String t is generated by randomly shuffling string s and then adding one more letter at a random position. Return the letter that was added to t.

```python
def find_the_difference(string1, string2):
    """
    This function takes as input 2 strings. String1 and String2.
    String2 is enerated by randomly shuffling string String1 
    and then adding one more letter at a random position. 
    This function returns the letter that was added to String1.
    """
    
    try:
        if not len(string1) == len(string2) - 1:
            print("lengths not as expected")
            return None
        
        s1 = set(string1)
        s2 = set(string2)
        
        if len(s2.difference(s1)) > 1:
            print("difference more than 1 byte")
            return None
        
        same = []
        diff = []
        sorted_string1 = sorted(string1)
        sorted_string2 = sorted(string2)
         
        i = 0
        j = 0
        
        while i < len(string1):
            if sorted_string1[i] == sorted_string2[j]:
               
                same.append(sorted_string1[i])
                
            else:
                
                diff.append(sorted_string2[j])
                j += 1
                if sorted_string1[i] == sorted_string2[j]:
                    
                    same.append(sorted_string1[i])

            i += 1
            j += 1

        if j == len(sorted_string2) - 1:
        #if len(diff) == 0:
            #print("vvv")
            diff.append(sorted_string2[j])

        #print(diff)
        if len(diff) == 1:
            return diff[0]
        else:
            print("added characters not equal to 1")
            return None
        
    except Exception as e:
        print("Check logs for error")
        lg.error("Error occured here")
        lg.exception(e)
```

### 2. Given a function that accepts unlimited arguments, check and count how many data types are in those arguments. Finally return the total in a list.

```python
def count_datatypes( *args ):
    """
    This function accepts any number of arguments, and checks and counts
    how many data types are in those arguments. It returns a list with the 
    number of elements of each data type at the corresponding position in the
    given data type list.
    """
    
    try:
        lg.info("Input by user: {}".format(args))
        
        list_order = [int, str, bool, list, tuple, dict]
        
        return_list = [0 for i in range(len(list_order))]
        
        for i in args:
            if type(i) in list_order:
                element_idx = list_order.index(type(i))
                return_list[element_idx] += 1
        
        return return_list
        
    except Exception as e:
        print("Check logs for error")
        lg.error("Error occured here")
        lg.exception(e)
```

```python
count_datatypes([10, 20], ("t", "Ok"), 2, 3, 1)
```

```python
count_datatypes("Hello", "Bye", True, True, False, {"1": "One", "2": "Two"}, [1, 3], {"Brayan": 18}, 25, 23)
```

### 3. A Fibonacci string is a precedence of the Fibonacci series. It works with any two characters of the English alphabet (as opposed to the numbers 0 and 1 in the Fibonacci series) as the initial items and concatenates them together as it progresses in a similar fashion as the Fibonacci series.

```python
def fib_str(num, list_of_chars):
    """
    This function takes as input:
    num - which is the number of elements needed in the Fibonacci string series
    list_of_chars - list of 2 characters which are required to start the Fibonacci string series
    
    The function returns the Fibonacci string with num as number of elements.
    """
    
    try:
        lg.info("Input by user, num: {}, list_of_chars: {}".format(num, list_of_chars))
        
        if not(len(list_of_chars)) == 2:
            print("there should be 2 characters given as list")
            return None
        
        if num < 1:
            print("num should be 1 or more")
        
        if num == 1:
            return [list_of_chars[0]]
        
        if num == 2:
            return list_of_chars
        
        return_list = [i for i in list_of_chars]
        
        i = num
        a = list_of_chars[0]
        b = list_of_chars[1]
        
        while i > 2:
            c = b + a
            a = b
            b = c
            
            return_list.append(c)
            
            i -= 1
        
        return return_list
        
    except Exception as e:
        print("Check logs for error")
        lg.error("Error occured here")
        lg.exception(e)
```

### 4. Given an integer between 0 and 26, make a variable (self.answer). That variable would be assigned to a string in this format:

"nines:your answer, threes:your answer, ones:your answer"

### You need to find out how many ones, threes, and nines it would at least take for the number of each to add up to the given integer when multiplied by one, three or nine (depends).

```python
def ones_threes_nines(num):
    """
    This funcition takes a number as input and return a string in the fomrat:
    "nines:your answer, threes:your answer, ones:your answer"
    where these values come from the calculationw which add up to give the number.
    """
    
    try:
        
        help_dict = {
                        'one': 1,
                        'two': 2,
                        'three': 3,
                        'four': 4,
                        'five': 5,
                        'six': 6,
                        'seven': 7,
                        'eight': 8,
                        'nine': 9
                    }

        string_format = "nines:your answer, threes:your answer, ones:your answer"
        
        num_list1 = re.findall(r"[a-z]+s:", string_format)
        num_list = []
        
        for i in range(len(num_list1)):
            num_list.append(num_list1[i].split("s:")[0])
            
        num_list_numeric = [help_dict[i] for i in num_list]
        
        count_list = [0 for i in range(len(num_list_numeric))]
        
        remainder = num
        
        for i in range(len(num_list_numeric)):
            count_list[i] = int( remainder / num_list_numeric[i] )
            remainder = remainder % num_list_numeric[i]
        
        
        return_string = ""
        
        for i in range(len(num_list1)):
            return_string += num_list1[i]
            return_string += " "
            return_string += str(count_list[i])
            return_string += ", "
        
        return_string = re.split(", $", return_string)[0]
        
        return return_string
        
    except Exception as e:
        print("Check logs for error")
        lg.error("Error occured here")
        lg.exception(e)
```

### 5. The Fibonacci sequence is a classic use case for recursive functions since the value of the sequence at a given index is dependent on the sum of the last two values. However, the recursion tree created by solving the Fibonacci sequence recursively can grow quite fast. Therefore it can be important to think about the implications of running a function recursively. Depending on the size of n needed and the capabilities of the system in question you might want to take a different approach.

### Write a non-recursive function that takes an integer n and returns the Fibonacci sequence's value at index n.

```python
def fib(num):
    """
    This function returns the value of nth element in the Fibonacci series
    """
    
    try:
        lg.info("Input by user: {}".format(num))
        
        a = 0
        b = 1
        
        if num == 0:
            return a
        
        n = num
        
        while n > 1:
            c = a + b
            a = b
            b = c
            
            n = n - 1
        
        return b
            
        
    except Exception as e:
        print("Check logs for error")
        lg.error("Error occured here")
        lg.exception(e)
```
